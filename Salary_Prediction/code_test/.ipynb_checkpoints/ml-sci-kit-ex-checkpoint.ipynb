### Scaling
#### What?
Adjust features so they are all roughly in the same range without losing their magnitude.

#### Why?
We can improve the performance of many machine learning algorithms. The specific reason differs depending on the task in hand:
<br><br>
**Regression**
* Most regression models are trained using gradient descent algorithm. This is just a way to minimise some objective function attached to the model.[Visual Representation](https://www.youtube.com/watch?v=kJgx2RcJKZY)
* The reason we standardise is to make each dimension on the plane more proportional - it would take longer to walk down the same gradient hill which was 100 miles vs 1 mile long.

**Clustering**
* The reason we normalise here is to standardise the distance between the clusters - comparing bannas and oranges.
* Some cases we don't want to do this - if we were comparing orange size in New Zealand and Hawaii - it wouldn't make sense to normalise.

#### How?
*Strategy 1:* Get the zero mean and unit variance of features

```python
from sklearn import preprocessing
import numpy as np
import pprint as pp

X_train = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.],
                    [ 1.,  -1., -2.]])

# We can "save" the transformation so we can revert back later
scaler = preprocessing.StandardScaler().fit(X_train)
# For note: features represent each column for scikit
# X1  X2  X3
# x11 x12 x13
# x21 x22 x23
# x31 x32 x33

# Transformed Matrix
pp.pprint(scaler.transform(X_train))
# Each features mean is 0
pp.pprint(scaler.transform(X_train).mean(axis=0))
# Each features variance is 1
pp.pprint(scaler.transform(X_train).std(axis=0))
# Can reverse the transformation later
pp.pprint(scaler.inverse_transform(scaler.transform(X_train)))
```

*Streatgy 2:* Scale to a range - normally [0,1]. This can be useful however it can constrain the variables too much compared to strategy 1.

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0,1))
X_train_minmax = min_max_scaler.fit_transform(X_train)
X_train_minmax
```

More techniques on scaling / normalisation and comparisons on them can be found [here](http://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#sphx-glr-auto-examples-preprocessing-plot-all-scaling-py) in scikit-learn docs.

### Support Vector Machines (classificaiton)
**Why** - Linear isn't good enough.

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
# use Seaborn plotting defaults
import seaborn as sns; sns.set()
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# use Seaborn plotting defaults
import seaborn as sns; sns.set()

from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');

xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)
for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')
    plt.xlim(-1, 3.5);
```

Here we can see there are 3 linear classifies which fit the data perfectly well. However the prediction of X would return different outputs from the classifies.

**What?** - Maximising the Margin
<br>
The main aim is to try and maximise the margin between the two points that are closest to the classifier. 

```python
# 1. Choose the model
from sklearn.svm import SVC
```

* Now we can also choose a hyperparameter for the SVM - in this it's represented as C. 
* This value allows us to soften the margin - essentially saying it's ok if a few points cross it if this allows for a better fit. 
* Here we have quiet spread data so we can be strict and set the value to be very large.

```python
# 2. Choose the hyperparameter
model = SVC(kernel='linear', C=1E10)
model.fit(X, y)
```

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a two-dimensional SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
        levels=[-1, 0, 1], alpha=0.5,
        linestyles=['--', '-', '--'])
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
            model.support_vectors_[:, 1],
            s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model);
```

#### Why are SVMs good?
**1. Very simple to define**
The points that touch the line above are known as the _support vectors_ and essentially define the model. You can retrieve them as such:

```python
model.support_vectors_
```

**2. Works very well for data with high dimensions, low samples and isn't affected by distant points**
Because it only has to work out the boundaires between classes - so doesn't need that much information the rest of the space.

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
def plot_svm(N=10, ax=None):
    X, y = make_blobs(n_samples=200, centers=2,
    random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=1E10)
    model.fit(X, y)
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [60, 120]):
    plot_svm(N, axi)
    axi.set_title('N = {0}'.format(N))
```

**3. Very flexible as you can use different kernels to adapt them to many different data types**
If you have a more non-linear boundary - you can use a kernel (polynomial and radial are two most popular ones) to transform the data and use support vector machine. This approach is called the kernel trick.

### Decision Tree / CART (classification)
**Why?**
Very intuitive algorithm to read - essentially you can arrive at a prediction by just reading through a set of pre-defined rules.

**How?**

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

# 4 classes across 2 dimensions
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=4,
random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

```python
# 1. Choose the model
from sklearn.tree import DecisionTreeClassifier
# 2. Choose hyperparameters for the model and fit
tree = DecisionTreeClassifier().fit(X, y)

# Helper function:
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()
    # Plot the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=cmap,
        clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    # fit the estimator
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
    np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
        levels=np.arange(n_classes + 1) - 0.5,
        cmap=cmap, clim=(y.min(), y.max()),
        zorder=1)
    ax.set(xlim=xlim, ylim=ylim)
```

The depper the tree the more complex boundairies we can have...

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)
for axi, N in zip(ax, [3, 10]):
    visualize_classifier(DecisionTreeClassifier(max_depth=N), X, y, axi)
```

But we can go to deep and start just fitting to that specific data - rather than the actual world. If we split the data in two we can see this - it learns patterns that are very specific to the data it is given.

### K-means Clustering
It searches for a predefined number of clusters/groups within the data using the expectation-maximisation (EM) algorithm to achieve this. This algorithm guesses some initial clusters and then:
1. (E-step) Assigns points to their nearest cluster center 
2. (M-step) Sets the cluster centres to the mean of all points in the current cluster. 
Repeats 1 and 2 until it converges.

K-means is very regularly applied to doing customer segmentaitons - so trying to define different customer types based on some behaviour defined very the data (frequency of purchase, average spend, products purchased...)

### Principle Component Analysis
* Aims to reduce the number of dimensions that "define" a dataset / set of variables. 
* It can be used on it's own to create latent / more abstracted variables or in combination with supervised algorithms to improve it's performance (through a reduction in input variables needed to be trained)

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
```

```python
# 1. Choose the model
from sklearn.decomposition import PCA
# 2. Choose the hyperparamters - components to reduce to
pca = PCA(n_components=1)
# 3. Train the model
pca.fit(X)
# 4. "predict"
X_pca = pca.transform(X)
print("original shape: ", X.shape)
print("transformed shape:", X_pca.shape)
```

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
X_new = pca.inverse_transform(X_pca)
plt.scatter(X[:, 0], X[:, 1], alpha=0.2)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha=0.8)
plt.axis('equal');
```

### Choosing the number of components: explained variance
* In the above example we only two variables so if we wanted to do dimensionality reduction we knew we were looking to reduce the dimensions to 1. 
* For higher dimensions - it won't be so obvious how many components to choose. 
* There is a trade-off between the number of components we choose and the explained variance in the data. This trade-off can be visually depicted using the PCA package.

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape # 64 features across a sample set of 1797
```

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

**What number of components do we choose?** 
Well it depends on the problem: 
* If we're trying to use PCA to get a better visual representation of the data - then really 2 to 3 (and at most 4) dimensions is the most we could have. 
* If we're just looking to reduce the dimensions to put into our machine learning model - then perhaps we'd look for some value where the slope starts to level out.

**Before** we do this we need to determine how to measure the accuracy of our models.

## What metrics can I use to measure the performance of my model?
Above we had a classificaiton problem so it can be relatively easy to to work out the model's performance (% of classes predicited correctly). For regression we need something else.


**RMSE** - root mean square error:
<br><br>
$RMSE = \sqrt{\frac{\sum_{i=1}^n (ypred_{i} - yactual_{i})^2}{n}}$
<br><br>
* Standard metric that be applied across all types of regression problems. 
* The squared term it penalises more heavily for predictions which are very different to the actual.

**MAPE** - mean absolute percentage error:
<br><br>
$MAPE = \frac{100}{n}\sum_{i=1}^n|\frac{yactual_{i} - ypred_{i}}{yactual_{i}}|$
<br><br>
*Can* be useful in a business context - gives a percentage error for the model. It has short-comings though:
1. If your actual value is zero you'll end up trying to divide by zero
2. Imbalanced: When forecast < actual the percentage error is bounded by 100% - but for the reverse this value would be unbounded.

```python
yactual= 100
ypred = 50
mape = abs((yactual-ypred)/yactual)
mape
```

#### Method 3: Cross - validation

* Splitting testing and training data like that is good but it now means the model doesn't see a good portion of the data. 
* For small samples of data this could be highly problematic as we might be missing out on a lot of information we could train the model on.

**Solution**: Cross-validation

Do multiple accuracy test on the data - splitting the training and testing data differently each time. Now we can simultaneously: 
1. Expose the model to all segments of the data (important to note each model is still only trained on a segment of the data at a time)
2. Understand it's global performance - how does it perform when trained and tested on different sets of the data. If we see a set of scores with high average and low variance across all trials we can be confident we have a good model.

### 1. Algorithm: Choose another algorithm

### 2. Hyperparamters: Choose better hyperparamters (grid search)
* There can be different hyperparamters which we need to initialise the model with before we can start to train. The selection of these can impact model performance significantly.
* Rather than manually trying out different hyperparamters we can use scikit learns *grid_search* function to automatically do this for us and select the optimal model (across a number of cross-validation sets).

```python
# Grid search
from sklearn.model_selection import GridSearchCV
# Here we can give it a range of hyperparamters to test the model with
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 1000]}
model = SVC() # don't initalise any paramters now
grid = GridSearchCV(model, parameters, cv=7, n_jobs=-1)
grid.fit(X,y)
grid.best_params_
```

GridSearch exhaustively tries every combination of hyperparameter to see which one fits best. This can very computationally expensive - so to get around this you can:
1. Use *n_jobs* to run across multiple cores
2. Use smarter searches offered by scikit learn (*RandomisedSearchCV*).

### 3. Feature Selection: Use a different number of features (feature selection)
Choosing the correct set of features can drastically improve the performance of our model. There are multiple strategies for doing this: <br><br>
#### S1: Removing features with low variance
If it doesn't vary that much why include it?

```python
from sklearn.feature_selection import VarianceThreshold
X = [[0, 2, 7], [0, 3, 6], [0, 4, 5], [0, 5, 4], [0, 6, 3], [0, 7, 2]]
sel = VarianceThreshold(threshold=0.15)
print('X: \n {} \n X_transform: \n {}'.format(X, sel.fit_transform(X)))
# Remove the first column as none
```

#### S2: Univariate Feature Selection
* Univariate (meaning of one variable) feature selection looks at the relationship between the target variable (so not useful for unsupervised learning) and each input variable individually. 
* Those that are considered to have a low relationships can be thrown from the model.
* Selection can be determine by rank (lowest 10 variables get kicked out) or by percentile. See more on this and metrics [here](http://scikit-learn.org/stable/modules/feature_selection.html#univariate-feature-selection).

```python
# Quick example:
X, y = iris.data, iris.target
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# Use chi2 method to get the top 2 variables in terms of relationship score
X_new = SelectKBest(chi2, k=2).fit_transform(X, y)
print('X: \n {} \n X_new: \n {}'.format(X[1:10], X_new[1:10]))
```

#### S3 Recursive Feature Elimination
* We build a model with all the features - get the feature importance of each of them and remove the one which has lowest importance. 
* This method is repeated until it reaches a specified number of features left. 
* In contrast to univariate selection previously - the feature importance statistics is in-built within the model being used. So the metric will vary from model to model.

```python
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
# Quick example:
from sklearn.svm import SVC
from sklearn.datasets import load_digits
from sklearn.feature_selection import RFE
# from sklearn.feature_selection import RFECV
import matplotlib.pyplot as plt

digits = load_digits()
# print(digits['images'])
X = digits.images.reshape((len(digits.images), -1))
y = digits.target

svc = SVC(kernel="linear", C=1)
# Define that we want to use the svc estimator, we'd like to go until the final feature and
# eliminate one features at a time.
rfe = RFE(estimator=svc, n_features_to_select=1, step=1)
# Fit data to RFE rather than SVC
rfe.fit(X, y)
ranking = rfe.ranking_.reshape(digits.images[0].shape)

# Plot pixel ranking
plt.matshow(ranking, cmap=plt.cm.Blues)
plt.colorbar()
plt.title("Ranking of pixels with RFE")
plt.show()
```

#### S4: Determine the actual impact of the variable on the target
* In S2 and S3 we've been looking at the statistical significance between the input variables and the target variable. 
* This is different to lookinga at the actual impact the variable has on the output.
* We could have two features that are very statistically significant but a unit change in one could have a much bigger impact on the target variable than the other.
* Loosely we can call this *Key Driver Analysis (KDA)*


*Example:*
* A client comes and say to you "how much does the weather affect my sales"?
* You could build a model with weather and other variables (pricing, seasonality, location of the store) to predict revenues
* Then using KDA you could try and work out the impact of weather on sales.

**Why?**
This method is great because it can go some way to understanding the impact of a single variable in highly non-linear model - normally such reasoning would apply to a linear regression model where we could just read the coefficents.

One of the main methods for doing KDA is called _partial dependance_. It works as such:
1. We choose the feature we wish to study and then hold that feature at a constant (say x1 = 1)
2. Iterate over all the other features in the model - recording the target value everytime.
3. Get the average of all the target values in 2. and associate it to when x1=1
4. Increment x1 (say x1=2) then repeat step 2. and 3. - and now assigning the average of step 2 to x1=2.
5. Repeat this for all values of x1.
<br>
The result will be a set of x1 values and with each value being assigned to it an _average target value_ for that x1 value. We can plot the results and see how a change in x1 affects the target value on average.

Scikit implementation can be found [here](http://scikit-learn.org/stable/auto_examples/ensemble/plot_partial_dependence.html)

```python
from __future__ import print_function
print(__doc__)
matplotlib.rcParams['figure.figsize'] = [15.0, 10.0]
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble.partial_dependence import plot_partial_dependence
from sklearn.ensemble.partial_dependence import partial_dependence
from sklearn.datasets.california_housing import fetch_california_housing

cal_housing = fetch_california_housing()

# split 80/20 train-test
X_train, X_test, y_train, y_test = train_test_split(cal_housing.data,
                                                    cal_housing.target,
                                                    test_size=0.2,
                                                    random_state=1)
names = cal_housing.feature_names

print("Training GBRT...")
clf = GradientBoostingRegressor(n_estimators=100, max_depth=4,
                                learning_rate=0.1, loss='huber',
                                random_state=1)
clf.fit(X_train, y_train)
print(" done.")

print('Convenience plot with ``partial_dependence_plots``')

features = [0, 5, 1, 2, (5, 1)]
fig, axs = plot_partial_dependence(clf, X_train, features,
                                   feature_names=names,
                                   n_jobs=3, grid_resolution=50)
fig.suptitle('Partial dependence of house value on nonlocation features\n'
             'for the California housing dataset')
plt.subplots_adjust(top=0.9)  # tight_layout causes overlap with suptitle

print('Custom 3d plot via ``partial_dependence``')
fig = plt.figure()

target_feature = (1, 5)
pdp, axes = partial_dependence(clf, target_feature,
                               X=X_train, grid_resolution=50)
XX, YY = np.meshgrid(axes[0], axes[1])
Z = pdp[0].reshape(list(map(np.size, axes))).T
ax = Axes3D(fig)
surf = ax.plot_surface(XX, YY, Z, rstride=1, cstride=1,
                       cmap=plt.cm.BuPu, edgecolor='k')
ax.set_xlabel(names[target_feature[0]])
ax.set_ylabel(names[target_feature[1]])
ax.set_zlabel('Partial dependence')
#  pretty init view
ax.view_init(elev=22, azim=122)
plt.colorbar(surf)
plt.suptitle('Partial dependence of house value on median\n'
             'age and average occupancy')
plt.subplots_adjust(top=0.9)

plt.show()
```

#### Lastly pipelining: joing it all together
We've seen above that there can be a lot of steps involved to processing the data, training the model and evaluation it's output (even when using scikit learn's pretty lean modules). We can essentially join these stages together into a pipeline and make our lives even easier. Below we do a quick example of doing PCA on a feature set and then feeding that into a logistic regression.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model, decomposition, datasets
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

# Define a logisitc regression model
logistic = linear_model.LogisticRegression()

# Define PCA
pca = decomposition.PCA()
# Build a pipeline with these 2 components
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])

# Load the data
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target

# Set out a range for our hyperparamter optimisation
n_components = [20, 40, 64]
Cs = np.logspace(-4, 4, 3)
# Load them into a dictionary
parameters = dict(pca__n_components=n_components,logistic__C=Cs)

# Here in one line we can:
# 1. Run the pipeline (of PCA and logistic regression) 
# 2. Do a GridSearch to find optimal hyperparameters
# 3. While also performing cross-validaiton
estimator = GridSearchCV(estimator = pipe, param_grid = parameters, cv=5 )
estimator.fit(X_digits, y_digits)
```

```python
print(estimator.best_params_)
```
