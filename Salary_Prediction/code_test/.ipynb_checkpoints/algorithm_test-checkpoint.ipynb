## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_estimators=1000, random_state=50, n_jobs=-1, min_samples_leaf=1)
forest.fit(X_data,Y_data.flatten())
```

```python
cross_val_score(forest, X_data, Y_data.flatten(), cv=10)
```

```python
r_score = forest.score(X_data, Y_data)
r_score
```
